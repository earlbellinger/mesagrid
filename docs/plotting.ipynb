# Plotting

Visualizing MESA models with mesagrid

```python
from mesagrid import Track, star_plots
```

```python
import matplotlib.pyplot as plt
plt.style.use('thinfig')
```

```python
track = Track('work')
```

## Hertzsprung-Russell Diagram

```python
star_plots.plot_hr(track,
    show_profiles=True,
    solar_symbol=True)
```

## Internal chemical composition

```python
star_plots.plot_composition(track, 5) # 5th profile file
```

## Propagation diagram

```python
star_plots.plot_propagation(track, 5)
```

## Echelle diagram

```python
star_plots.plot_echelle(track, 5)
```

## Interactive plotting

```python
star_plots.star_interact(track)
```
