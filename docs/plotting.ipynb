# Plotting

Visualizing MESA models with mesagrid

```python
from mesagrid import Track, star_plots
```

```python
import matplotlib.pyplot as plt
plt.style.use('thinfig')
```

```python
track = Track('work')
```

```python
star_plots.plot_hr(track)
```

```python
star_plots.plot_composition(track, 5)
```

```python
star_plots.plot_propagation(track, 5)
```

```python
star_plots.plot_echelle(track, 5)
```
